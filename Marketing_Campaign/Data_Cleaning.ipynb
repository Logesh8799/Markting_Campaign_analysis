Convert Dt_Customer to date type and derive Customer_Tenure

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce')
reference_date = pd.to_datetime('2026-07-01')
df['Customer_Tenure_Days'] = (reference_date - df['Dt_Customer']).dt.days
df['Customer_Tenure_Months'] = (reference_date - df['Dt_Customer']).dt.days // 30
print(df[['Dt_Customer', 'Customer_Tenure_Days', 'Customer_Tenure_Months']].head())
```

Age from Year_Birth

```python
from datetime import datetime
#current_year = datetime.datetime.now().year
current_year = datetime.today().year

df['Age'] = current_year - df['Year_Birth']

print(df[['Year_Birth', 'Age']].head())
```

Total_Spend as sum of all Mnt* fields.

```python
spend_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Spend'] = df[spend_columns].sum(axis=1)
print(df[spend_columns + ['Total_Spend']].head())
```

Total_Purchases as sum of channel purchases.

```python
channel_columns = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']
df['Total_Purchases'] = df[channel_columns].sum(axis=1)
print(df[channel_columns + ['Total_Purchases']].head())
```

Childrens at home

```python
children_columns = ['Kidhome', 'Teenhome']
df['Total_Children'] = df[children_columns].sum(axis=1) 
print(df[children_columns + ['Total_Children']].head())
```

Outliers handling on Total_Spend

```python
column_name = 'Total_Spend'

Q1 = df[column_name].quantile(0.25)
Q3 = df[column_name].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]

print(f"Lower limit for normal data: {lower_bound}")
print(f"Upper limit for normal data: {upper_bound}")
print(f"Number of outliers found: {len(outliers)}")
```

```python
df['Total_Spend'] = df['Total_Spend'].clip(lower=lower_bound, upper=upper_bound)

print("Outliers capped! Max spend is now:", df['Total_Spend'].max())
```

```python
import numpy as np
import pandas as pd

# 1. Load your dataset
df = pd.read_csv(r"C:\Users\My Lap\OneDrive\Desktop\Marketing_Campaign\cleaned_data.csv")

# 2. Calculate the 90th percentile threshold dynamically for High Spenders
high_spend_threshold = df["Total_Spend"].quantile(0.90)

# ==========================================
# Apply the exact text values based on rules
# ==========================================

# 1. High Income (Shows the range text or "Not High")
df["Segment_High_Income"] = np.where(df["Income"] > 75000, "75001–99999", "Normal Income")

# 2. Young Customer (Shows the age range text or "30+")
df["Segment_Young_Customer"] = np.where(df["Age"] < 30, "18–29", "30+")

# 3. Campaign Responder (Shows "1" or "0")
df["Segment_Campaign_Responder"] = np.where(df["Response"] == 1, "1", "0")

# 4. High Web Engagement (Shows the visit range text or "<=5")
df["Segment_High_Web_Engagement"] = np.where(df["NumWebVisitsMonth"] > 5, "6–20", "0–5")

# 5. Family Customer (Shows the children range text or "0")
df["Segment_Family_Customer"] = np.where(df["Total_Children"] > 0, "1–3", "0")

# 6. High Spender (Shows "> ₹50,000" or "Normal Spender")
df["Segment_High_Spender"] = np.where(df["Total_Spend"] > high_spend_threshold, f"> ₹{int(high_spend_threshold):,}", "Normal Spender")


# 3. Save the segmented data to a new CSV file
df.to_csv("segmented_data.csv", index=False)

# 4. Preview the results to verify the new text values
preview_cols = [col for col in df.columns if col.startswith("Segment_")]
print(df[preview_cols].head(10))
```
